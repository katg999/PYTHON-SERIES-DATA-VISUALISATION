GROUP BY AND AGGREGATE FUNCTIONS

```python
import pandas as pd
df = pd.read_csv('/content/Flavors.csv')
df
```

```python
group_by_frame = df.groupby('Flavor')
```

```python
group_by_frame.count()
```

```python
group_by_frame.sum()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean', 'max', 'count', 'sum'], 'Texture Rating': ['mean', 'max', 'count', 'sum']})
```

Merge, Join and Concatenate

```python
import pandas as pd
df1 = pd.read_csv('/content/LOTR.csv')
df1
```

```python
import pandas as pd
df2 = pd.read_csv('/content/LOTR 2.csv')
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='right')
```

```python
df1.merge(df2, how='cross')
```

```python
df1.join(df2, on = 'FellowshipID', how='outer', lsuffix='_Left', rsuffix='_Right')
```
